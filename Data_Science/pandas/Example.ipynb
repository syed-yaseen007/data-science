```python
import pandas as pd
import time
```

```python
df = pd.read_csv("Example.csv")
```

```python
df
```

```python
# Boolean indexing
df[(df["Department"] == "IT") & (df["Salary"] > 65000)]

# query
df.query('Department == "IT" and Salary > 65000')
```

```python
df["Join Date"] = pd.to_datetime(df["Join Date"])

# Boolean indexing
df[df["Join Date"] > "2022-01-01"]

# query
df.query('`Join Date` > "2022-01-01"')
```

```python
# Active employees from Mumbai or Delhi.
df[(df["Status"]=="Active")&(df["City"]=="Mumbai"|"Delhi")]
df.query("'Status'=='Active'and 'City'=='Mumbai'or 'Delhi'")
```

```python
# Employees younger than 30 in HR.
df.query("'Department'=='HR'>30")
df[(df["Department"=='HR'])&(df['age'>30])]
```
